## Prepare The DataSet For Training

```python
#Perform the train-test split for that sales segment  
train_len = 500
train = data[0:train_len] 
test = data[train_len:] 
```

```python
print(test)
```

```python
from pylab import rcParams
import statsmodels.api as sm
rcParams['figure.figsize'] = 12, 8
decomposition = sm.tsa.seasonal_decompose(train.sales_amount, model='additive') # additive seasonal index

fig = decomposition.plot()
plt.show()
```

```python
decomposition = sm.tsa.seasonal_decompose(train.sales_amount, model='multiplicative') # multiplicative seasonal index
fig = decomposition.plot()
plt.show()
```

```python
y_hat_naive = test.copy()
y_hat_naive['naive_forecast'] = train['sales_amount'][train_len-1] #Last observation in the train data
```

```python
plt.figure(figsize=(12,4))
plt.plot(train['sales_amount'], label='Train')
plt.plot(test['sales_amount'], label='Test')
plt.plot(y_hat_naive['naive_forecast'], label='Naive forecast')
plt.legend(loc='best')
plt.title('Naive Method', fontweight= 'bold')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(test['sales_amount'], y_hat_naive['naive_forecast'])).round(2)
mape = np.round(np.mean(np.abs(test['sales_amount']-y_hat_naive['naive_forecast'])/test['sales_amount'])*100,2)

results = pd.DataFrame({'Method':['Naive method'], 'MAPE': [mape], 'RMSE': [rmse]})
results = results[['Method', 'RMSE', 'MAPE']]
results
```

## Simple Average Method

```python
y_hat_avg = test.copy()
y_hat_avg['avg_forecast'] = train['sales_amount'].mean()# Average of past months sales
```

```python
plt.figure(figsize=(12,4))
plt.plot(train['sales_amount'], label='Train')
plt.plot(test['sales_amount'], label='Test')
plt.plot(y_hat_avg['avg_forecast'], label='Simple average forecast')
plt.legend(loc='best')
plt.title('Simple Average Method', fontweight='bold')
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(test['sales_amount'], y_hat_avg['avg_forecast'])).round(2)
mape = np.round(np.mean(np.abs(test['sales_amount']-y_hat_avg['avg_forecast'])/test['sales_amount'])*100,2)

tempResults = pd.DataFrame({'Method':['Simple average method'], 'RMSE': [rmse],'MAPE': [mape] })
results = pd.concat([results, tempResults])
results = results[['Method', 'RMSE', 'MAPE']]
results
```

## Simple moving average forecast

```python
y_hat_sma = data.copy()
ma_window = 3  # Considered 3 month window
y_hat_sma['sma_forecast'] = data['sales_amount'].rolling(ma_window).mean()
y_hat_sma['sma_forecast'][train_len:] = y_hat_sma['sales_amount'][train_len-1]
```

```python
plt.figure(figsize=(12,4))
plt.plot(train['sales_amount'], label='Train')
plt.plot(test['sales_amount'], label='Test')
plt.plot(y_hat_sma['sma_forecast'], label='Simple moving average forecast')
plt.legend(loc='best')
plt.title('Simple Moving Average Method', fontweight= 'bold')
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(test['sales_amount'], y_hat_sma['sma_forecast'][train_len:])).round(2)
mape = np.round(np.mean(np.abs(test['sales_amount']-y_hat_sma['sma_forecast'][train_len:])/test['sales_amount'])*100,2)

tempResults = pd.DataFrame({'Method':['Simple moving average forecast'], 'RMSE': [rmse],'MAPE': [mape] })
results = pd.concat([results, tempResults])
results = results[['Method', 'RMSE', 'MAPE']]
results
```

## Simple exponential smoothing forecast

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
model = SimpleExpSmoothing(train['sales_amount'])
model_fit = model.fit(optimized=True)
model_fit.params
y_hat_ses = test.copy()
y_hat_ses['ses_forecast'] = model_fit.forecast(6)
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
model = ExponentialSmoothing(np.asarray(train['sales_amount']) ,seasonal_periods=12 ,trend='additive', seasonal=None)
model_fit = model.fit(smoothing_level=0.2, smoothing_slope=0.01, optimized=False)
print(model_fit.params)
y_hat_holt = test.copy()
y_hat_holt['holt_forecast'] = model_fit.forecast(len(test))
```

```python
plt.figure(figsize=(12,4))
plt.plot(train['sales_amount'], label='Train')
plt.plot(test['sales_amount'], label='Test')
plt.plot(y_hat_ses['ses_forecast'], label='Simple exponential smoothing forecast')
plt.legend(loc='best')
plt.title('Simple Exponential Smoothing Method')
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(test['sales_amount'], y_hat_holt['holt_forecast'])).round(2)
mape = np.round(np.mean(np.abs(test['sales_amount']-y_hat_holt['holt_forecast'])/test['sales_amount'])*100,2)

tempResults = pd.DataFrame({'Method':['Holt\'s exponential smoothing method'], 'RMSE': [rmse],'MAPE': [mape] })
results = pd.concat([results, tempResults])
results = results[['Method', 'RMSE', 'MAPE']]
results
```

```python
from scipy.stats import boxcox
data_boxcox = pd.Series(boxcox(data['sales_amount'], lmbda=0), index = data.index)

plt.figure(figsize=(12,4))
plt.plot(data_boxcox, label='After Box Cox tranformation')
plt.legend(loc='best')
plt.title('After Box Cox transform', fontweight= 'bold')
plt.show()
```

```python
data_boxcox_diff = pd.Series(data_boxcox - data_boxcox.shift(), data.index)
plt.figure(figsize=(12,4))
plt.plot(data_boxcox_diff, label='After Box Cox tranformation and differencing')
plt.legend(loc='best')
plt.title('After Box Cox transform and differencing', fontweight='bold')
plt.show()
```
